```python
import os
import shutil
import re

def organize_examples(root_dir):
    # Regular expression pattern to extract mood, genre, and bpm from the filename
    pattern = re.compile(r'mood=([A-Za-z]+)_genre=([A-Za-z]+)_BPM=(\d+)_.*\.')

    # Iterate through each subdirectory in the root directory
    for subfolder in os.listdir(root_dir):
        subfolder_path = os.path.join(root_dir, subfolder)

        # Check if it is a directory
        if os.path.isdir(subfolder_path):
            # Create a dictionary to store files based on example name
            example_files = {}

            # Iterate through files in the subdirectory
            for filename in os.listdir(subfolder_path):
                file_path = os.path.join(subfolder_path, filename)
                if not os.path.isfile(file_path):
                    continue

                # Use regex to extract mood, genre, and bpm from the filename
                match = pattern.match(filename)

                if match:
                    mood, genre, bpm = match.groups()

                    # Create example name using the extracted values
                    example_name = f"{mood}-{genre}-{bpm}"

                    # Create a subsubfolder for the example if it doesn't exist
                    example_folder = os.path.join(subfolder_path, example_name)
                    os.makedirs(example_folder, exist_ok=True)

                    # Copy the file to the example subsubfolder
                    shutil.copy(file_path, example_folder)

    print("Organization complete.")



# Replace 'your_root_directory' with the path to your root directory containing subfolders
organize_examples('Demo_Assets_Cherrypicked')
```

```python
import os
import ffmpeg

def convert_wavs_to_mp3s(root_dir):
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            if file.endswith('.wav'):
                wav_path = os.path.join(subdir, file)
                mp3_path = os.path.splitext(wav_path)[0] + '.mp3'
                ffmpeg.input(wav_path).output(mp3_path, **{'qscale:v': 1}).overwrite_output().run()

# Replace 'your_root_directory' with the path to your root directory containing WAVs
convert_wavs_to_mp3s('Demo_Assets_Cherrypicked')
```

```python
import os
import shutil

def create_trash_folder(root_dir):
    trash_folder = os.path.join(root_dir, 'trash')
    os.makedirs(trash_folder, exist_ok=True)
    return trash_folder

def move_non_png_mp3_to_trash(root_dir, trash_folder):
    for subdir, _, files in os.walk(root_dir):
        for file in files:
            file_path = os.path.join(subdir, file)

            # Check if the file is not a PNG or MP3
            if not (file.lower().endswith('.png') or file.lower().endswith('.mp3')):
                # Move the file to the trash folder
                shutil.move(file_path, os.path.join(trash_folder, file))

# Replace 'your_root_directory' with the path to your root directory
root_directory = 'Demo_Assets_Random'

# Create 'trash' folder
trash_folder = create_trash_folder(root_directory)

# Move non-PNG and non-MP3 files to 'trash' folder
move_non_png_mp3_to_trash(root_directory, trash_folder)
```
